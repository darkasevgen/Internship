```python
import numpy as np
import pandas as pd
import os

import io
from datetime import datetime
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('new.csv')
df
```

```python
df['date'] = df['date'].astype(str)
df['time'] = df['time'].astype(str)
```

```python
df['date2'] = df['date'] + ' ' + df['time']
df
```

```python
df['datetime'] = pd.to_datetime(df.date2, format='%Y%m%d %H%M%S')
```

```python
df = df[['datetime', 'price']]
```

```python
df
```

```python
df = df.set_index('datetime')
```

# Стратегия 1

```python
df[df.index <= '2015-07-01'].shape[0]
```

```python
df[df.index <= '2015-02-01'].shape[0]
```

2 скользящих средних - одна экпоненциальная (31 день), другая - обычная, 182-183 дня

```python
df['ma_31exp'] = df.price.ewm(span=df[df.index <= '2015-02-01'].shape[0], adjust=False).mean()
df['ma_182'] = df.price.rolling(window=df[df.index <= '2015-07-01'].shape[0], min_periods=1).mean()
```

```python
df.plot(figsize=(20, 8), grid=True)
plt.axvline('2015-07-01', c='r')
plt.show()
```

```python
df['diff'] = df['ma_31exp'] - df['ma_182']
```

```python
u = np.sign(df['diff'])
m = np.flatnonzero(u.diff().abs().eq(2))
```

Индексы смены знака между Exp31 и MA182 (если быстрая МА, пробила медленную => покупать; если при открытом ордере пробила обратно => закрывать)

```python
m
```

```python
df2 = df.iloc[m][df.iloc[m].index > '2015-07-01']
df2
```

```python
profit = df2.loc['2016-10-05 18:39:00', 'price'] - df2.loc['2015-10-30 15:29:00', 'price']
profit
```

Т.е. покупка за 997 и продажа за 1367

```python
df[['price', 'ma_31exp', 'ma_182']].plot(figsize=(20, 8), grid=True, zorder=1)
plt.axvline('2015-07-01', c='r')
plt.scatter('2015-10-30 15:29:00', 997.0, marker='^', s=100, c='r', zorder=2)
plt.scatter('2016-10-05 18:39:00', 1367.0, marker='v', s=100, c='r', zorder=2)
plt.show()
```

# Стратегия 2

В задании упоминается, что сделки только купить-продать, а не наоборот

```python
df = df[['price']]
```

```python
df['ma_30'] = df.price.rolling(window=df[df.index <= '2015-02-01'].shape[0], min_periods=1).mean()
df['ma_273'] = df.price.rolling(window=df[df.index <= '2015-09-01'].shape[0], min_periods=1).mean()
```

```python
df.plot(figsize=(20, 8), grid=True)
plt.axvline('2015-09-01', c='r')
plt.show()
```

При пробитии медленной МА => покупать

При сильном ОБРАТНОМ пробое (70 пунктов) быстрой МА => продавать

```python
df[df.index > '2015-09-01'].plot(figsize=(30, 8), grid=True, zorder=1)
# plt.scatter('2015-12-07 15:29:00', 997.0, marker='^', s=100, c='r', zorder=2)
plt.show()
```

```python
df['diff_ma_price'] = df['ma_30'] - df['price']
```

```python
df = df[df.index > '2015-09-01']
```

```python
df2 = df[(df['diff_ma_price'] > 70) & (df['ma_273'] < df['price'])]
```

```python
plt.subplots(figsize=(20, 8))
plt.plot(df.index, df.price, zorder=1)
plt.scatter(df2.index, df2.price, c='r', s=5, zorder=2)
plt.show()
```

В области февраля 2016 акции как раз-таки будут куплены (т.к. медленная МА будет пробита ценой, после этого следующий ОБРАТНЫЙ пробой быстрой МА произойдёт в ~ июле 2016)

```python
df2 = df[(df['diff_ma_price'] > 70) & (df['ma_273'] < df['price'])]
df2
```

```python
df3 = df[(df['ma_273'] < df['price'])]
df3
```

Первая транзакция

```python
df[['price', 'ma_30', 'ma_273']][df.index > '2015-09-01'].plot(figsize=(30, 8), grid=True, zorder=1)
plt.scatter(df3.index[0], df3.price[0], marker='^', s=100, c='r', zorder=2)
plt.scatter(df2.index[0], df2.price[0], marker='v', s=100, c='r', zorder=2)
plt.show()
```

```python
profit2 = df2.price[0] - df3.price[0]
profit2
```

Вторая транзакция

```python
df5 = df2[df2.index > '2016-05']
df5
```

```python
df4 = df[(df['ma_273'] > df['price']) & (df.index > '2016-01-20') & (df.index < '2016-01-21')]
df4
```

Беру последний индекс пока цена держится ниже МА

```python
df4.index[-1]
```

```python
df.loc[df4.index[-1]:]
```

Видно, что цена в 2016-01-21 10:01:00 пробивает МА

```python
df[['price', 'ma_30', 'ma_273']][df.index > '2015-09-01'].plot(figsize=(30, 8), grid=True, zorder=1)
plt.scatter(df3.index[0], df3.price[0], marker='^', s=100, c='r', zorder=2)
plt.scatter(df2.index[0], df2.price[0], marker='v', s=100, c='r', zorder=2)
plt.scatter(df.loc[df4.index[-1]:].index[1], df.loc[df4.index[-1]:].price[1], marker='^', s=100, c='r', zorder=2)
plt.scatter(df5.index[0], df5.price[0], marker='v', s=100, c='r', zorder=2)
plt.show()
```

```python
profit2 += (df5.price[0] - df.loc[df4.index[-1]:].price[1])
```
